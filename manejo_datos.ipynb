# Manejo de datos

Cargar la librería de pandas y el archivo desde GitHub (path)

```python
import pandas as pd
path = "https://raw.githubusercontent.com/IsacNM/manejo_datos/5ec3b0da06765ff4a649a339a5c2fd89ecdb6479/pipol_dataset.csv"
df = pd.read_csv(path)
```

 Información del Dataframe

```python
print("INFO DEL DATAFRAME:")
df.info()
```

Primeros 5 elementos

```python
print("PRIMEROS 5 ELEMENTOS (head):")
print(df.head())
```

Mostrar columnas: Nombre y País

```python
print("COLUMNAS: Nombre y País")
print(df[["name", "country"]])
```

Mostrar 3 filas por índice (series)

```python
print("FILAS POR ÍNDICE (0, 1, 2) - Series:")
print("Índice 0:")
print(df.iloc[0])
print("\nÍndice 1:")
print(df.iloc[1])
print("\nÍndice 2:")
print(df.iloc[2])
```

Filtrar: todos los que pertenecen a Méxicoprint("\n" + "=" * 50)

```python
print("FILTRO: Personas de México")
mexico = df[df["country"] == "Mexico"]
print(mexico)
```

Filtrar: México y que viven en Dusty City

```python
print("FILTRO: México y Dusty City")
mexico_dusty = df[(df["country"] == "Mexico") & (df["city"] == "Dusty City")]
print(mexico_dusty)
```
